```python
#Aplicação para manipulação de números

import random

lista_num = []

def criar_lista_aleatoria(N):
    global lista_num
    lista_num = [random.randint(1, 100) for _ in range(N)]
    print("Lista aleatória criada com sucesso!")

def ler_lista():
    global lista_num
    N = int(input("Introduza o total de números: "))
    lista_num = []
    for _ in range(N):
        lista_num.append(int(input("Introduza um número: ")))
    print("Lista criada com sucesso!")

def calcular_soma():
    soma = sum(lista_num)
    print(f"Soma dos elementos da lista: {soma}")

def calcular_media():
    media = sum(lista_num) / len(lista_num)
    print(f"Média dos elementos da lista: {media}")

def encontrar_maior():
    maior = max(lista_num)
    print(f"O maior número da lista é: {maior}")

def encontrar_menor():
    menor = min(lista_num)
    print(f"O menor número da lista é: {menor}")

def crescente():
    print("A lista está organizada por ordem crescente?")
    ordenada_crescente = sorted(lista_num)
    if lista_num == ordenada_crescente:
        print("Sim!")
    else:
        print("Não")

def decrescente():
    print("A lista está organizada em ordem decrescente?")
    ordenada_decrescente = sorted(lista_num, reverse=True)
    if lista_num == ordenada_decrescente:
        print("Sim!")
    else:
        print("Não")

def procurar_elemento(elemento):
    if elemento in lista_num:
        posicao = lista_num.index(elemento)
        print(f"O elemento {elemento} está na posição {posicao}.")
    else:
        print(f"O elemento {elemento} não está na lista.")

while True:
    print("\nMenu:")
    print("1- Criar Lista Aleatória")
    print("2- Criar Lista Manualmente")
    print("3- Soma")
    print("4- Média")
    print("5- Maior")
    print("6- Menor")
    print("7- Está organizada por ordem crescente?")
    print("8- Está organizada por ordem decrescente?")
    print("9- Procurar um Elemento")
    print("0- Sair")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        criar_lista_aleatoria(int(input("Introduza o tamanho da lista: ")))

    elif opcao == 2:
        ler_lista()

    elif opcao == 3:
        calcular_soma()
        input("Pressione Enter para continuar...")

    elif opcao == 4:
        calcular_media()
        input("Pressione Enter para continuar...")

    elif opcao == 5:
        encontrar_maior()
        input("Pressione Enter para continuar...")

    elif opcao == 6:
        encontrar_menor()
        input("Pressione Enter para continuar...")

    elif opcao == 7:
        crescente()
        input("Pressione Enter para continuar...")

    elif opcao == 8:
        decrescente()
        input("Pressione Enter para continuar...")

    elif opcao == 9:
        elemento = input("Introduza o elemento que deseja procurar: ")
        procurar_elemento(elemento)
        input("Pressione Enter para continuar...")

    elif opcao == 0:
        print("A aplicação irá encerrar!")
        print(f"Lista atual: {lista_num}")
        break
```
